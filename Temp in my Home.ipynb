```python
df = pd.read_csv(r"/content/drive/MyDrive/IoT Data - Sheet1.tsv", sep='\t')
df.head()
```

```python
# Check for missing values
df.isnull().sum()
# Fill missing values or drop rows as needed
df = df.dropna()  # Drop rows with missing data
```

```python
# Get a summary of statistics for numerical columns
df.describe()
```

```python
# Check correlation between temperature and humidity
df[['Temperature', 'Humidity']].corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot temperature over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Timestamp', y='Temperature', data=df)
plt.title('Temperature Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.show()

# Plot humidity over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Timestamp', y='Humidity', data=df)
plt.title('Humidity Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Humidity (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot of average temperature by timestamp (or any categorical column)
plt.figure(figsize=(10, 6))
sns.barplot(x='Timestamp', y='Temperature', data=df)
plt.title('Average Temperature by Timestamp')
plt.xlabel('Timestamp')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Temperature vs. Humidity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature', y='Humidity', data=df)
plt.title('Temperature vs. Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
# Box plot of Temperature
plt.figure(figsize=(10, 6))
sns.boxplot(x='Temperature', data=df)
plt.title('Temperature Distribution (Box Plot)')
plt.xlabel('Temperature (°C)')
plt.show()

# Box plot of Humidity
plt.figure(figsize=(10, 6))
sns.boxplot(x='Humidity', data=df)
plt.title('Humidity Distribution (Box Plot)')
plt.xlabel('Humidity (%)')
plt.show()
```

```python
# Calculate the average of Temperature
average_temperature = df['Temperature'].mean()

print(f"Average Temperature: {average_temperature} °C")
```

```python
# Calculate the average of Humidity
average_humidity = df['Humidity'].mean()

print(f"Average Humidity: {average_humidity} %")
```

```python
# Calculate the minimum temperature
min_temperature = df['Temperature'].min()

print(f"Minimum Temperature: {min_temperature} °C")
```
